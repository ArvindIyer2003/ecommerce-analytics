```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("=" * 70)
print("OLIST E-COMMERCE DATA EXPLORATION")
print("=" * 70)

# File paths
data_path = '../data/raw/'

# Load all datasets
print("\nğŸ“‚ Loading datasets...")

try:
    orders = pd.read_csv(f'{data_path}olist_orders_dataset.csv')
    print("âœ“ Orders loaded")
except Exception as e:
    print(f"âœ— Orders failed: {e}")
    
try:
    order_items = pd.read_csv(f'{data_path}olist_order_items_dataset.csv')
    print("âœ“ Order Items loaded")
except Exception as e:
    print(f"âœ— Order Items failed: {e}")

try:
    customers = pd.read_csv(f'{data_path}olist_customers_dataset.csv')
    print("âœ“ Customers loaded")
except Exception as e:
    print(f"âœ— Customers failed: {e}")

try:
    products = pd.read_csv(f'{data_path}olist_products_dataset.csv')
    print("âœ“ Products loaded")
except Exception as e:
    print(f"âœ— Products failed: {e}")

try:
    sellers = pd.read_csv(f'{data_path}olist_sellers_dataset.csv')
    print("âœ“ Sellers loaded")
except Exception as e:
    print(f"âœ— Sellers failed: {e}")

try:
    payments = pd.read_csv(f'{data_path}olist_order_payments_dataset.csv')
    print("âœ“ Payments loaded")
except Exception as e:
    print(f"âœ— Payments failed: {e}")

try:
    reviews = pd.read_csv(f'{data_path}olist_order_reviews_dataset.csv')
    print("âœ“ Reviews loaded")
except Exception as e:
    print(f"âœ— Reviews failed: {e}")

try:
    geolocation = pd.read_csv(f'{data_path}olist_geolocation_dataset.csv')
    print("âœ“ Geolocation loaded")
except Exception as e:
    print(f"âœ— Geolocation failed: {e}")

try:
    category_translation = pd.read_csv(f'{data_path}product_category_name_translation.csv')
    print("âœ“ Category Translation loaded")
except Exception as e:
    print(f"âœ— Category Translation failed: {e}")

print("\nâœ“ All datasets loaded successfully!")
```

```python
print("\n" + "=" * 70)
print("DATASET OVERVIEW")
print("=" * 70)

datasets = {
    'Orders': orders,
    'Order Items': order_items,
    'Customers': customers,
    'Products': products,
    'Sellers': sellers,
    'Payments': payments,
    'Reviews': reviews,
    'Geolocation': geolocation,
    'Category Translation': category_translation
}

for name, df in datasets.items():
    print(f"\nğŸ“Š {name.upper()}")
    print(f"   Rows: {len(df):,}")
    print(f"   Columns: {len(df.columns)}")
    print(f"   Memory: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    print(f"   Columns: {', '.join(df.columns[:5])}..." if len(df.columns) > 5 else f"   Columns: {', '.join(df.columns)}")
```

```python
print("\n" + "=" * 70)
print("DATA QUALITY ANALYSIS")
print("=" * 70)

for name, df in datasets.items():
    print(f"\nğŸ” {name.upper()}")
    
    # Null counts
    nulls = df.isnull().sum()
    if nulls.sum() > 0:
        print(f"   âš ï¸  Null values found:")
        for col, count in nulls[nulls > 0].items():
            pct = (count / len(df)) * 100
            print(f"      - {col}: {count:,} ({pct:.1f}%)")
    else:
        print("   âœ“ No null values")
    
    # Duplicates
    dupes = df.duplicated().sum()
    if dupes > 0:
        print(f"   âš ï¸  Duplicate rows: {dupes:,}")
    else:
        print("   âœ“ No duplicate rows")
```

```python
print("\n" + "=" * 70)
print("TABLE RELATIONSHIPS")
print("=" * 70)

# Orders to Order Items (1-to-many)
print("\nğŸ”— ORDERS â†” ORDER ITEMS")
print(f"   Unique orders in orders table: {orders['order_id'].nunique():,}")
print(f"   Unique orders in order_items table: {order_items['order_id'].nunique():,}")
print(f"   Total order line items: {len(order_items):,}")
avg_items = len(order_items) / orders['order_id'].nunique()
print(f"   Average items per order: {avg_items:.2f}")

# Orders to Customers (many-to-1)
print("\nğŸ”— ORDERS â†” CUSTOMERS")
print(f"   Unique customers in orders: {orders['customer_id'].nunique():,}")
print(f"   Unique customers in customer table: {customers['customer_id'].nunique():,}")
orders_per_customer = len(orders) / orders['customer_id'].nunique()
print(f"   Average orders per customer: {orders_per_customer:.2f}")

# Order Items to Products (many-to-1)
print("\nğŸ”— ORDER ITEMS â†” PRODUCTS")
print(f"   Unique products in order_items: {order_items['product_id'].nunique():,}")
print(f"   Unique products in products table: {products['product_id'].nunique():,}")

# Order Items to Sellers (many-to-1)
print("\nğŸ”— ORDER ITEMS â†” SELLERS")
print(f"   Unique sellers in order_items: {order_items['seller_id'].nunique():,}")
print(f"   Unique sellers in sellers table: {sellers['seller_id'].nunique():,}")

# Orders to Payments (1-to-many, some orders have multiple payments)
print("\nğŸ”— ORDERS â†” PAYMENTS")
print(f"   Unique orders in payments: {payments['order_id'].nunique():,}")
payment_methods = payments.groupby('order_id').size()
print(f"   Max payment methods per order: {payment_methods.max()}")
print(f"   Payment types: {payments['payment_type'].unique()}")

# Orders to Reviews (1-to-1)
print("\nğŸ”— ORDERS â†” REVIEWS")
print(f"   Unique orders in reviews: {reviews['order_id'].nunique():,}")
print(f"   Review coverage: {(reviews['order_id'].nunique() / orders['order_id'].nunique()) * 100:.1f}%")
```

```python
print("\n" + "=" * 70)
print("BUSINESS INSIGHTS PREVIEW")
print("=" * 70)

# Date range
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
print(f"\nğŸ“… DATE RANGE")
print(f"   First order: {orders['order_purchase_timestamp'].min()}")
print(f"   Last order: {orders['order_purchase_timestamp'].max()}")
print(f"   Time span: {(orders['order_purchase_timestamp'].max() - orders['order_purchase_timestamp'].min()).days} days")

# Revenue
print(f"\nğŸ’° REVENUE")
total_revenue = payments['payment_value'].sum()
print(f"   Total revenue: R$ {total_revenue:,.2f}")
avg_order_value = payments.groupby('order_id')['payment_value'].sum().mean()
print(f"   Average order value: R$ {avg_order_value:,.2f}")

# Product categories
print(f"\nğŸ“¦ PRODUCTS")
print(f"   Unique categories: {products['product_category_name'].nunique()}")
print(f"   Top 5 categories:")
category_counts = products['product_category_name'].value_counts().head()
for cat, count in category_counts.items():
    print(f"      - {cat}: {count}")

# Geography
print(f"\nğŸŒ GEOGRAPHY")
print(f"   Unique customer states: {customers['customer_state'].nunique()}")
print(f"   Top 5 states:")
state_counts = customers['customer_state'].value_counts().head()
for state, count in state_counts.items():
    print(f"      - {state}: {count:,} customers")

# Order status
print(f"\nğŸ“Š ORDER STATUS")
status_counts = orders['order_status'].value_counts()
for status, count in status_counts.items():
    pct = (count / len(orders)) * 100
    print(f"   {status}: {count:,} ({pct:.1f}%)")
```
